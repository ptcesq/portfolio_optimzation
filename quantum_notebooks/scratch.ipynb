```python
def generate_all_binary_portfolios(N_Assets: int):
    """
    Generate all possible binary portfolios x in {0,1}^N_Assets.
    """
    for Binary_Decision_Vector in product([0, 1], repeat=N_Assets):
        yield np.array(Binary_Decision_Vector, dtype=int)
        return Binary_Decision_Vector 

N_Assets = 4 
Total_Portfolios = 2 ** N_Assets
print(f"Total number of portfolios to evaluate: {Total_Portfolios}")
```
